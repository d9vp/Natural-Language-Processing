```python
# Import Libraries
import nltk
from nltk.corpus import wordnet as wn
nltk.download('wordnet')
```

```python
# Identify Word Senses using Synsets in NLTK
word = "bank"
synsets = wn.synsets(word)
```

```python
# Print the Synsets and their Definitions
print("The word '{}' has {} synsets in WordNet:".format(word, len(synsets)))
for synset in synsets:
    print("- Synset '{}' -> Definition: {}".format(synset.name(), synset.definition()))
```

Thus, we can see that bank has 10 senses in its noun form and as a verb, it has 7.
