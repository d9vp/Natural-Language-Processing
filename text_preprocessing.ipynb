The Natural Language Toolkit (NLTK) is a platform used for building Python programs that work with human language data for applying in statistical natural language processing (NLP).

### Remove Numbers


```python
def remove_numbers(text):
  result = re.sub(r'\d+', '', text)
  return result
  
input_str = "There are 3 balls in this bag, and 12 in the other one."
print(remove_numbers(input_str) )
```

### Remove Punctuation


```python
def remove_punctuation(text):
  translator = str.maketrans('', '', string.punctuation)
  return text.translate(translator)

input_str = "Hey, did you know that the summer break is coming? Amazing right !! It's only 5 more days !!"
print(remove_punctuation(input_str))
```

### Remove Whitespaces


```python
def remove_whitespace(text):
  return " ".join(text.split())
  
input_str = " we don't need   the given       questions"
print(remove_whitespace(input_str) )
```

### Remove Stop Words

```python
from nltk.corpus import stopwords
```

```python
def remove_stopwords(text):
  stop_words = set(stopwords.words("english"))
  word_tokens = word_tokenize(text)
  filtered_text = [word for word in word_tokens if word not in stop_words]
  return filtered_text

text="we do not need the given questions"
print(remove_stopwords(text))
```

### Lemmatization

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
lemmatizer = WordNetLemmatizer()
# lemmatize string
def lemmatize_word(text):
  word_tokens = word_tokenize(text)
  # provide context i.e. part-of-speech
  lemmas = [lemmatizer.lemmatize(word, pos ='v') for word in word_tokens]
  return lemmas
text = 'data science uses scientific methods algorithms and many types of processes'
print(lemmatize_word(text) )
```

###Word Frequency Count

```python
from collections import Counter
def word_count(Words):
  my_counter = Counter()
  for word in Words:
      my_counter.update(word.split())
  return my_counter
word_count(['This', 'counts', 'how', 'frequenty', 'words', 'appear', 'in', 'this', 'corpus', 'filled', 'with', 'words'])
```

```python
from num2words import num2words
def numtowords(text):
  text=text.split(' ')
  rettext=''
  for i in text:
    if i.isnumeric():
      rettext+= num2words(i)+' '
    elif '.' in i:
      ind=i.index('.')
      if i[:ind].isnumeric() and i[ind+1:].isnumeric():
        rettext+= num2words(i)+' '
    else:
      rettext+= i+' '
  return rettext
numtowords('There were 13 people with a 9.5 score')
```

##Dataset

```python
import pandas as pd
data=pd.read_csv('/content/sample_data/tweets.csv')
df=data[['text']]
df.head()
```

```python
words=[]
lowercase_text=[]
wo_no=[]
wo_punc=[]
wo_ws=[]
wo_sw=[]
lemma=[]
stems=[]
wc=[]
for i in range(len(df)):
  lowercase_text.append(df['text'][i].lower())        #Converting text to lowercase
  wo_no.append(remove_numbers(lowercase_text[i]))     #Removing numbers from output text 
  wo_punc.append(remove_punctuation(wo_no[i]))        #Removing punctuation from output text
  wo_ws.append(remove_whitespace(wo_punc[i]))         #Removing whitespaces from output text
  wo_sw.append(remove_stopwords(wo_ws[i]))            #Removing stopwords from output text
  temp=[]
  for j in wo_sw[i]:                                  #Tokenise the output text
    temp.append(word_tokenize(j)[0])
  stems.append(temp)
  temp=[]
  for j in stems[i]:                                  #Lemmatizing the words obtained after tokenisation
    temp.append(lemmatize_word(j)[0])
  lemma.append(temp)
  wc.append(word_count(lemma[i]))
df['Lowercase']=lowercase_text
df['Without numbers']=wo_no
df['Without Punctuation']=wo_punc
df['Without Whitespaces']=wo_ws
df['Without Stop Words']=wo_sw
df['Stemming']=stems
df['Lemmatization']=lemma
df['Word Count']=wc
df.head()
```
