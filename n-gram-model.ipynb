```python
import nltk
from nltk.util import ngrams
from nltk import word_tokenize
import pandas as pd
import re
import string
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
def text_lowercase(text):
  return text.lower()
def remove_numbers(text):
  result = re.sub(r'\d+', '', text)
  return result
def remove_punctuation(text):
  return re.sub(r'[^\w\s]', '', text)
def remove_whitespace(text):
  return " ".join(text.split())
def remove_stopwords(text):
  stop_words = set(stopwords.words("english"))
  word_tokens = word_tokenize(text)
  filtered_text = [word for word in word_tokens if word not in stop_words]
  return filtered_text
def lemmatize_word(text):
  word_tokens = word_tokenize(text)
  lemmas = [lemmatizer.lemmatize(word, pos ='v') for word in word_tokens]
  return lemmas
```

```python
f1 = open("/content/sample_data/national_science_day.txt", "r")
data = f1.read()
f1.close()
```

```python
text=text_lowercase(data)
text=remove_numbers(text)
text=remove_punctuation(text)
text=remove_whitespace(text)
text=remove_stopwords(text)
temp=[]
for i in text:                                  
  temp.append(word_tokenize(i))
text=temp
temp=[]
for i in text:     
  temp.append((lemmatize_word(i[0]))[0])                         
text=temp
data=''
for i in text:
  data=data+i+" "
```

```python
def extract_ngrams(data, num):
  n_grams = ngrams(nltk.word_tokenize(data), num)
  return [ ' '.join(grams) for grams in n_grams]

print("2-gram: ", extract_ngrams(data, 2))
print("3-gram: ", extract_ngrams(data, 3))
print("4-gram: ", extract_ngrams(data, 4))
```

```python
text = data.lower()
words = text.split()

words = [word.strip(".,!;()[]”’‘“-'`") for word in words]
fin = [word.replace("’s", '') for word in words]
unique = []
for word in words:
    if word not in unique:
        unique.append(word)
unique.extend(['','s'])
print(unique)
print(len(unique))
```

```python
empty=[[0 for i in range(len(unique))] for j in range(len(unique))]
df = pd.DataFrame(empty,columns=unique,index=unique)
print(df.head())
```

```python
#compute frequency distribution for all the bigrams in the text
bgs=extract_ngrams(text, 2)
fdist = nltk.FreqDist(bgs)
print(fdist)
for k,v in fdist.items():
  print(k,v)
```

```python
for k,v in fdist.items():
  temp=k.strip(".,!;()[]’”‘'`“-")
  temp=temp.replace("’s", '')
  temp=temp.split(" ")
  temp1=temp[0]
  temp2=temp[1]
  df[temp1][temp2]=v
  # print(v)
print(df)
```

```python
bgs=extract_ngrams(text, 1)
fdist = nltk.FreqDist(bgs)
print(fdist)
unigram=dict()
for k,v in fdist.items():
  unigram.update({k:v})
print(unigram)
```

```python
def getprob(df,uni,word):
  word=word.lower()
  listprob=[]
  for i in range(len(unique)):
    prob=(df[word][unique[i]])/uni[word]
    listprob.append((unique[i],prob))
  for i in listprob:
    if i[1]>0:
      print(i)
  fin=max(listprob,key=lambda item:item[1])
  print(fin)
  print(f"Most probable word after '{word}' is:",fin[0])
```

```python
getprob(df,unigram,'science')
```
