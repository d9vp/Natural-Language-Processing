```python
vowel=['a','e','i','o','u']
conso=['b','c','d','f','g','h','j','k','l','m','n','p','q','r','s','t','v','w','x','y','z']
```

```python
def getm(word):
  m=0
  for i in range(len(word)-1):
    if word[i] in vowel:
      if word[i+1] in conso:
        m+=1
    elif word[i] in conso:
      if word[i+1] =='y':
        if ((i+2)<len(word)):
          if word[i+2] in conso:
            m+=1
  # print("m =",m)
  return m
```

```python
def step1a(word):
  if word.endswith('sses'):
    word = word.replace('sses', 'ss')
  elif word.endswith('ies'):
    word = word.replace('ies', 'i')
  elif word.endswith('ss'):
    word = word.replace('ss', 'ss')
  elif word.endswith('s'):
    word = word.replace('s', '')
  return word
```

```python
def step1b23(word):
  if word.endswith('at') or word.endswith('bl') or word.endswith('iz'):
    return word+'e'
  elif word.endswith('s'):
    return word[:-1]
  elif word[-1]==word[-2]:
    if word[-1] not in ['l','s','z']:
      return word[:-1]
  elif getm(word)==1:
    if word[-3] in conso:
      if word[-2] in vowel:
        if((word[-1] in conso) and (word[-1] not in ['w','x','y'])):
          return word+'e'
  return word
def step1b(word):
  flag=0
  if word.endswith('eed'):
    ind=word.rfind('eed')
    temp=word[:ind]
    m=getm(temp)
    if m>0:
      temp+='ee'
    return temp
  elif word.endswith('ed'):
    ind=word.rfind('ed')
    temp=word[:ind]
    for i in temp:
      if i in vowel:
        flag=1
    if flag==1:
      word=step1b23(temp)
      return word
  elif word.endswith('ing'):
    flag=0
    ind=word.rfind('ing')
    temp=word[:ind]
    for i in temp:
      if i in vowel:
        flag=1
    if flag==1:
      word=step1b23(temp)
      return word
  elif word.endswith('s'):
    return word[:-1]
  return(word)
```

```python
def step1c(word):
  flag=0
  if word[-1]=='y':
    for i in word[:-1]:
      if i in vowel:
        flag=1
        break
    if flag==1:
      return word[:-1]+'i'
  return word
```
